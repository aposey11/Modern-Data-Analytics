```python
from meteostat import Stations

# Rough bounding box of Flanders
stations = Stations()
stations = stations.bounds((50.6, 2.5), (51.6, 5.9))  # (southwest, northeast)

df = stations.fetch()
print(df.head())
print(len(df))
```

```python
from meteostat import Point, Daily
from datetime import datetime

# coordinates the first in the list 
lat = 50.916183
lon = 4.456122

# location
location = Point(lat, lon)

# time range, another data (2019 was not given) just to check
start = datetime(2023, 1, 1)
end = datetime(2023, 1, 7)

# fetch weather
data = Daily(location, start, end).fetch()

# show full output
print(data)

# optional: only rain
print("\nRain (prcp):")
print(data["prcp"])
```
